```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv")
df.head(10)
```

```python
landing_outcomes = df['Outcome'].value_counts()
landing_outcomes
```

```python
# Enumerate es una funcion para iterar sobre una secuencia de lista
# Landing_outcomes tenemos un frame, convertimos en diccionario y extraemos una lista de las keys

for i, outcome in enumerate(landing_outcomes.keys()):
    print(i, outcome)
```

```python
# Otra forma mas comprimida agregando _ = para evitar lista de [None]
_ = [print(i, outcome) for i , outcome in enumerate(landing_outcomes.keys())]
```

```python
# Usamos set para agrupar junto a los indices que queremos
bad_outcomes = set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

```python
# usando la columna outcome, crearemos una lista de 0 y 1

landing_class = []

for outcome in df['Outcome']:
    if outcome in bad_outcomes:
        landing_class.append(0)
    else:   
        landing_class.append(1)

```

```python
landing_class = [0 if outcome in bad_outcomes else 1 for outcome in df['Outcome']]
```

```python
df = pd.read_csv('../../data/processed/p_car_data.csv')
df.head(5)
```

```python
data_comp_avg_price = df[['brand','price']].groupby('brand', as_index = False).mean().rename(columns={'price':'brand_avg_price'})

df = df.merge(data_comp_avg_price, on ='brand')


df['brand_category'] = df['brand_avg_price'].apply(lambda x: "Budget" if x < 10000
                                                   else ("Mid_Range" if 10000 <= x < 20000
                                                         else "Luxury"))

columns=['price', 'fueltype', 'aspiration','carbody', 'drivewheel','wheelbase', 'brand_category',
                  'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 'carlength','carwidth','citympg','highwaympg']


selected = df[columns]
```

```python
# selected.dtypes.items() seleccionamos los tipos de la columna y convertimos en dicc
# agrega la llave por cada llave , valor en el diccitioario si es igual a O
categorical_columns = [key for key, value in selected.dtypes.items() if value == 'O']
categorical_columns
```
