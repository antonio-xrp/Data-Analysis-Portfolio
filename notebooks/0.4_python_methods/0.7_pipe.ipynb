```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pre_df = pd.read_csv('../../data/interim/0.1_penguins.csv')
pre_df.head(4)
```

La funci√≤n pipe() sirve como apply(), la diferencia esta que pipe aplica funciones a todo el data frame, y apply solo a filas o columnas

```python
pre_df.isnull().melt(value_name='missing').pipe(lambda pre_df: (sns.displot(data = pre_df,
                                                                            y='variable',
                                                                            hue='missing',
                                                                            multiple='fill',
                                                                            aspect=2)))
```
