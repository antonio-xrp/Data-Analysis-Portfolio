```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('./SpaceX_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[['payloads','window']]

df.isnull().sum()
```

```python
details_mean = df['window'].mean()

# Usamos replace para remplazar valores faltantes con su media
df.loc[:,'window'] = df['window'].replace(np.nan, details_mean)
```

```python
df.isnull().sum()
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-ML240EN-SkillsNetwork/labs/data/CarPrice_Assignment.csv')
df.head(5)
```

```python
df.CarName.value_counts()
```

```python
# Primero cortaremos las palabras

df['brand'] = df.CarName.str.split(' ').str.get(0).str.lower()
df['brand'].unique()
```

```python
# Existe algunos typos

df['brand'] = df['brand'].replace(['vw', 'vokswagen'], 'volkswagen')
df['brand'] = df['brand'].replace(['maxda'], 'mazda')
df['brand'] = df['brand'].replace(['porcshce'], 'porsche')
df['brand'] = df['brand'].replace(['toyouta'], 'toyota')

df.brand.unique()
```
